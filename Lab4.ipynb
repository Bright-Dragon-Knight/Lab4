## 1. Getting started with Data Analysis-Importing the Dataset

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('rawautodata.csv',header = None)
df.head()
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df.columns = headers
df.head()
```

## 2. Data Pre-processing/Data Cleaning/Data Wrangling

```python
import numpy as np
```

```python
df.info()
```

```python
df.replace('?',np.nan,inplace=True)
df.head(n=10)
```

```python
missingdata = df.notnull()
missingdata.head(3)
```

```python
print(missingdata['normalized-losses'].value_counts())
```

```python
for column in missingdata.columns.values.tolist():
    print(column)
    print (missingdata[column].value_counts())
    print("") 
```

```python
mean_normloss = df['normalized-losses'].astype('float').mean(axis=0)
df['normalized-losses'].replace(np.nan,mean_normloss,inplace=True)
missingdata = df.notnull()
print(df['normalized-losses'].value_counts)
```

```python
list = ['stroke','bore','horsepower','peak-rpm']
for column in list:
    mean_value = df[column].astype('float').mean(axis=0)
    df[column].replace(np.nan,mean_value,inplace=True)
```

```python
from collections import Counter
Counter(df['num-of-doors'])
```

```python
df['num-of-doors'].replace(np.nan,'four',inplace=True)
```

```python
df['price'].dropna()
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.tail(10)
```

```python
for column in missingdata.columns.values.tolist():
    print(column)
    print(missingdata[column].value_counts())
    print('')
```

```python
df.to_csv('No_missing')
```

## Data Normalisation

```python
list2 = ['length','width','height']
for column in list2:
    df[column] = ((df[column] - df[column].min())
    / (df[column].max()-df[column].min()))
df[['length','width','height']].head()
```

```python
df['horsepower'].describe()
```

```python
df['horsepower'].nunique()
```

```python
df['horsepower'] = df['horsepower'].astype(int,copy=True)

import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(data = df,x ='horsepower' )
plt.title('Total Horsepower Bins')
plt.show()
```

## Data Normalisation

```python
bins = np.linspace(min(df['horsepower']),max(df['horsepower']),4)
group_names = ['low','midium','high']
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
sns.histplot(data = df,x ='horsepower' ,bins = 3)
plt.title('Total Horsepower Bins')
plt.show()
```

```python
dummy_fueltype = pd.get_dummies(df["fuel-type"])
dummy_fueltype
```

```python
dummy_fueltype.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)
df = pd.concat([df, dummy_fueltype], axis=1)
df.drop("fuel-type", axis = 1, inplace=True)
```

```python
df['aspiration'].unique()
```

```python
dummy_aspiration = pd.get_dummies(df['aspiration'])
dummy_aspiration.rename(columns={'std':'as_type_std','turbo':'as_type_turbo'},inplace=True)
df = pd.concat([df,dummy_aspiration],axis=1)
df.drop('aspiration',axis=1,inplace=True)
```

```python
df.to_csv('Normalization_Binning')
```
